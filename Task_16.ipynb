```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('./insurance.csv')
```

```python
# Q1.
print("First 10 Rows : ")
df.head(10)
```

```python
#Q2.
print("Shape of Dataset : ",df.shape)
print("\nDataset Information : ")
df.info()
print("\nStatistical summary : ")
df.describe()
```

```python
# Q3.
print("Mising Value : ")
df.isnull().sum()
```

```python
# Q4.
numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns
categorical_cols = df.select_dtypes(include=['object']).columns

print("\nNumerical Columns:")
print(list(numerical_cols))

print("\nCategorical Columns:")
print(list(categorical_cols))
```

```python
# Q5.

import matplotlib.pyplot as plt
import seaborn as sns
plt.subplot(2,2,1)
sns.histplot(df["age"])
plt.title("Age")
plt.xlabel("Age")
plt.ylabel("Count")

plt.subplot(2,2,2)
sns.histplot(df["bmi"])
plt.title("BMI")
plt.xlabel("BMI")
plt.ylabel("Count")

plt.subplot(2,2,3)
sns.histplot(df["children"])
plt.title("Children")
plt.xlabel("Children")
plt.ylabel("Count")

plt.subplot(2,2,4)
sns.histplot(df["expenses"])
plt.title("Expenses")
plt.xlabel("Expenses")
plt.ylabel("Count")
plt.tight_layout()
plt.show()
```

```python
# Q6
sns.countplot(x="sex", data=df)
plt.title("Sex")
plt.show()

sns.countplot(x="smoker", data=df)
plt.title("Smoker")
plt.show()

sns.countplot(x="region", data=df)
plt.title("Region")
plt.show()



```

```python
# Q7. Heatmap
import seaborn as sns
import matplotlib.pyplot as plt
corr_matrix = df.select_dtypes(include='number').corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5 )
plt.title("Correlation Heatmap of Numerical Variables")
plt.show()
```

```python
# Q8. Charges analysis
print("Average Expenses :", df["expenses"].mean())
print("Maximum Expenses :", df["expenses"].max())
print("Minimum Expenses :", df["expenses"].min())

print(df.groupby("smoker")["expenses"].mean())
```

```python
# Q9. Boxplots
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8, 6))

sns.boxplot(x='smoker', y='expenses', hue='sex', data=df, palette='Set2')

plt.title('Insurance expenses by Smoker and Sex')
plt.xlabel('Smoker')
plt.ylabel('expenses')
plt.legend(title='Sex')
plt.show()
```

```python
# 10.
print("\n******* Analysis Summary ******")
print("Average Age :", df["age"].mean())
print("Average BMI :", df["bmi"].mean())

print("\nRegion with Highest Customers:")
print(df["region"].value_counts())

print("\nObservation:")
print("* Smokers generally have much higher insurance charges.")
print("* BMI has a slight positive relationship with charges.")
print("* Age also increases insurance charges.")
print("* The region with the highest customer count is shown above.")
```
