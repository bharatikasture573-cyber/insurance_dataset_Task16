```python
# 4.
import matplotlib.pyplot as plt
import seaborn as sns
numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns
categorical_cols = df.select_dtypes(include=['object']).columns

print("\nNumerical Columns:")
print(list(numerical_cols))
print("\nCategorical Columns:")
print(list(categorical_cols))
```

```python
# 5
import matplotlib.pyplot as plt
import seaborn as sns
plt.subplot(2,2,1)
sns.histplot(df["car_id"])
plt.title("CAR ID")
plt.xlabel("Car_Id")
plt.ylabel("Count")

plt.subplot(2,2,2)
sns.histplot(df["price"])
plt.title("PRICE")
plt.xlabel("Price")
plt.ylabel("Count")

plt.subplot(2,2,3)
sns.histplot(df["year"])
plt.title("YEAR")
plt.xlabel("Year")
plt.ylabel("Count")

plt.subplot(2,2,4)
sns.histplot(df["user_rating"])
plt.title("RATING")
plt.xlabel("user_ratings")
plt.ylabel("Count")
plt.tight_layout()
plt.show()
```

```python
# 6
import matplotlib.pyplot as plt
import seaborn as sns
sns.countplot(x="brand", data=df)
plt.title("Brand")
plt.show()

sns.countplot(x="fuel_type", data=df)
plt.title("Fuel Type")
plt.show()

sns.countplot(x="color", data=df)
plt.title("Color")
plt.show()S
```

```python
# 7.
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

num_data = df[['car_id', 'price', 'year', 'accessories_cost', 'user_rating']]

corr = num_data.corr()

plt.figure(figsize=(8,6))
sns.heatmap(corr, annot=True, fmt=".2f")

plt.title("Correlation Heatmap of Numerical Variables")
plt.show()
```

```python
# 8.
import matplotlib.pyplot as plt
import seaborn as sns
print("Average Price :", df["price"].mean())
print("Maximum Price :", df["price"].max())
print("Minimum Price :", df["price"].min())
print("\nAverage Price by Fuel Type")
print(df.groupby("fuel_type")["price"].mean())
```

```python
# 9.

import matplotlib.pyplot as plt
import seaborn as sns

sns.boxplot(x="brand", y="price", data=df)
plt.title("Price by Brand")
plt.show()

sns.boxplot(x="fuel_type", y="price", data=df)
plt.title("Price by Fuel Type")
plt.show()
```

```python
# 10
print("\n************ ADVANCED ANALYSIS SUMMARY ************")
print(f"Average Car Price           : ₹{df['price'].mean():,.2f}")
print(f"Average Manufacturing Year  : {df['year'].mean():.2f}")
print(f"Average User Rating         : {df['user_rating'].mean():.2f}")

print(f"\nMost Expensive Car Price    : ₹{df['price'].max():,.2f}")
print(f"Least Expensive Car Price   : ₹{df['price'].min():,.2f}")

print("\nTop 5 Car Brands:")
print(df['brand'].value_counts().head())

corr = df[['price','year','accessories_cost','user_rating']].corr()
print("\nCorrelation Matrix:")
print(corr)

print("\n******* OBSERVATIONS *******")
print("1. The dataset contains both numerical and categorical features.")
print("2. The average car price gives an overview of the market value.")
print("3. Manufacturing year helps identify newer and older vehicles.")
print("4. User ratings indicate customer satisfaction with different cars.")
print("5. The most common brand has the highest number of listed cars.")
print("6. Correlation analysis shows how price changes with year, accessories cost, and user rating.")
print("7. Boxplots and histograms help identify outliers and data distribution.")
```
